```python
import requests
from bs4 import BeautifulSoup
```

```python
# Send an HTTP request to the webpage
url = 'https://www.popvortex.com/music/charts/top-100-songs.php'
page = requests.get(url)

# Parse the HTML response
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
title = []
artist = []

num_iter = len(soup.select("div.chart-content.col-xs-12.col-sm-8 > p .title"))

titles = soup.select("div.chart-content.col-xs-12.col-sm-8 > p .title")
artists = soup.select("div.chart-content.col-xs-12.col-sm-8 > p .artist")
# iterate through the result set and retrive all the data
for i in range(num_iter):
    title.append(titles[i].get_text())
    artist.append(artists[i].get_text())
print(title)
print(artist)
```

```python
import pandas as pd
# each list becomes a column
top100 = pd.DataFrame({"title":title,
                       "artist":artist
                      })
```

```python
import random
user_song = input("Enter a song title you like: ").lower()
user_artist = input("Enter the artist of the song you like: ").lower()

song_data = top100
song_data = song_data.applymap(lambda s: s.lower() if type(s) == str else s)

song_matches = song_data.loc[(song_data['title'] == user_song) & (song_data['artist'] == user_artist)]
song_artist = (song_data['artist'] == user_artist)

if song_matches.empty:
    print("We can not recommend a song.")
elif not song_artist.empty:
    indexes = top100.index
    random_index =random.choice(indexes)
    recommended_song = top100.iloc[random_index]['title']
    recommended_artist =top100.iloc[random_index]['artist']
    print("We recommend the following song: ", recommended_song, " by ", recommended_artist)
else:
    indexes = top100.index
    random_index =random.choice(indexes)
    recommended_song = top100.iloc[random_index]['title']
    recommended_artist =top100.iloc[random_index]['artist']
    print("We recommend the following song: ", recommended_song, " by ", recommended_artist)
```
